```python
datasets = {'mbpp_py' : 'mbpp_kendall.pickle',
            'he_py' : 'he_kendall.pickle',
            'he_go' : 'he_mupliple_go_corr.pickle',
            'he_rs' : 'he_mupliple_rs_corr.pickle',
            'he_java' : 'he_mupliple_java_corr.pickle'}
```

```python
data = {}
```

```python
for d in datasets:
    data[d] = pickle.load(open(datasets[d], 'rb'))
```

```python
total_Q = []

for i in range(len(data['he_py'])):
    
    q_list = []
    for d in data:
        q_list.append(data[d][i][0])

    #q += min(abs(data_mbpp[i][0]), abs(data_he[i][0]), abs(data_mpru[i][0]), abs(data_java[i][0]))
    min_v = np.sign(np.mean(q_list)) * min([abs(x) for x in q_list])
    total_Q.append((np.mean(q_list) + min_v, data['he_py'][i][1], data['he_py'][i][2]))
```
